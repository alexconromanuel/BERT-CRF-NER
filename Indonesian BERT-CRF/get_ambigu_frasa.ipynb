#Phrase

```python
import json
with open("/content/singgalang-test.json", encoding='UTF-8') as f:
    list_sents = json.load(f)
```

```python
list_tmp = []
for idx, x in enumerate(list_sents):
  for idy, y in enumerate(x.get('entities')):
    if len(y.get('text').split()) >1 :
      list_tmp.append(x)
      break
```

```python
list_tmp[:5]
```

```python
print(len(list_tmp))
print(len(list_sents))
```

```python
import codecs
with open("singgalang_phrase.json", "wb") as f:
    json.dump(list_tmp,
              codecs.getwriter("utf-8")(f),
              ensure_ascii=False,
              indent = 4)
```

#Ambigu

```python
import pandas as pd
import csv

dataset = pd.read_csv("https://raw.githubusercontent.com/ir-nlp-csui/singgalang/main/SINGGALANG.tsv", header = None, delimiter="\t", names=['token', 'entitas'], quoting=csv.QUOTE_NONE, encoding='utf-8')
```

```python
dataset['distinctentitas'] = dataset['token'].map(dataset.groupby('token')['entitas'].nunique(dropna = True))
```

```python
dataset.loc[dataset['entitas'] == 'Person'].head(30)
```

```python
dataset.loc[dataset['entitas'] == 'Organisation'].head(30)
```

```python
index_names = dataset[ dataset['distinctentitas'] == 1 ].index
dataset.drop(index_names, inplace = True)
```

```python
dataset = dataset.drop_duplicates()
```

```python
dataset.loc[dataset['entitas'] != 'O'].tail(30)
```

```python
list_token_ambigu = dataset['token'].values.tolist()
print(type(list_token_ambigu))
```

```python
list_token_ambigu[:5]
```

```python
def compare(l1, l2):
    # here l1 and l2 must be lists
    # if len(l1) != len(l2):
    #     return False
    # l1.sort()
    # l2.sort()
    if bool (set(l1) & set(l2)):
        return True
    else:
        return False
```

```python
lista_tmp = []
for idx, x in enumerate(list_sents):
  for idy, y in enumerate(x.get('entities')):
    print(compare(y.get('text').split(), list_token_ambigu))
    if compare(y.get('text').split(), list_token_ambigu) == True :
      lista_tmp.append(x)
      
      break
  break
```

```python
lista_tmp[0:]
```

```python
import json
with open("/content/ambigu.json", encoding='UTF-8') as f:
    list_ambigu = json.load(f)
```

```python
list_ambigu = list_ambigu+lista_tmp
```

```python
list_ambigu[10:]
```

```python
import codecs
with open("singgalang_ambigu.json", "wb") as f:
    json.dump(list_ambigu,
              codecs.getwriter("utf-8")(f),
              ensure_ascii=False,
              indent = 4)
```
